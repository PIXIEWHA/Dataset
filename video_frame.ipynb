```python
import cv2
import os
import glob

video = 'outsidedoor_07'

# 디렉토리 경로
path_dir = '/content/drive/MyDrive/trash_class/img/'
folder_list = os.listdir(path_dir+video)

print(folder_list)

# 디렉토리 안 영상들 프레임 추출
for i in folder_list:
  file_list = glob.glob('./'+video+'/'+i+'/*.mp4')

  for j in file_list:
    vidcap = cv2.VideoCapture(j)
    
    print(j+" 영상 총 프레임 수 : %f" %vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    count = 1
    success = True

    while success:
      success, image = vidcap.read()
      
      # 프레임
      frame = int(vidcap.get(1))
      if(frame % 60 == 0):

        # 이미지 사이즈 변경
        #image = cv2.resize(image, (960, 540))
        
        title = path_dir + video + '/' + os.path.basename(j)+"_%d_frame_%d.jpg" % (count, frame)
        cv2.imwrite(title, image)
        print(title+" 저장")
    
        if cv2.waitKey(10) == 27:                    
          break
        
        count += 1
```
