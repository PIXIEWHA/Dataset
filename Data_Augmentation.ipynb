```python
import numpy as np
import os
import cv2
import glob
from PIL import Image
import PIL.ImageOps
```

```python
file_path = '/content/drive/MyDrive/guide_trash/haein_aug/images'
file_names = os.listdir(file_path)
num_files = len(file_names)

for i in range(2000):
  print(file_names[i])
  file_name = file_names[i]

  original_path = file_path + '/' + file_name
  print(original_path)
  aug_path = '/content/drive/MyDrive/guide_trash/haein_aug/images_aug'

  img = cv2.imread(original_path)
  row, col, ch = img.shape
  mean = 0
  var = 0.1
  sigma = var**0.5
  gaussian = np.random.normal(mean, sigma, (row, col, ch))
  gaussian = gaussian.reshape(row, col, ch)
  noise_arr = img + gaussian
  noise_img = Image.fromarray(np.uint8(noise_arr)).convert('RGB')
  noise_img.save(aug_path + '/' + file_name + '_aug.jpg')
```
